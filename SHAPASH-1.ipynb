```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
from math import sqrt
```

```python
x_train=pd.read_excel('x_train.xlsx')
x_train=x_train.drop(['Unnamed: 0'],axis=1)
x_test=pd.read_excel('x_test.xlsx')
x_test=x_test.drop(['Unnamed: 0'],axis=1)
y_train=pd.read_excel('y_train.xlsx')
y_train=y_train.drop(['Unnamed: 0',],axis=1)
y_test=pd.read_excel('y_test.xlsx')
y_test=y_test.drop(['Unnamed: 0'],axis=1)
```

```python
from xgboost import XGBRegressor
xgbr = XGBRegressor(base_score=1, cv=4, learning_rate=0.15, max_depth=20,n_estimators=100, min_child_weight=3, objective='reg:squarederror', random_state=15, verbose=20 )
```

```python
xgbr.fit(x_train, y_train)
```

```python
base_score=1, cv=4, learning_rate=0.15, max_depth=20, min_child_weight=3, n_estimators=100, objective='reg:squarederror', random_state=15, verbose=20
```

```python
y_pred = pd.DataFrame(xgbr.predict(x_test),columns=['pred'], index=x_test.index)
```

```python
xgb_best_pred = xgbr.predict(x_test)
xgb_best_RMSE = sqrt(mean_squared_error(y_test, xgb_best_pred))
print(xgb_best_RMSE)
```

```python
xpl = SmartExplainer(xgbr)
xpl.compile(x=x_test,y_pred=y_pred)
```

```python
xpl.plot.features_importance()
```
